```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns

# Line plots
#Positive gradient
x =[1,2,3]
y =[1,2,3]

plt.plot(x,y)
```

```python
# Negative gradient
x =[1,2,3]
y =[3,2,1]
plt.plot(x,y)
```

```python

days = ['Moinday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday']
sales = [450,200,616,438,500,250]
plt.plot(days,sales, color='green',marker='*')  # o, D, v
plt.title('Line plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid()
plt.show()
```

```python
# Bar plots
#Positive gradient
```

```python
plt.bar(days,sales,color='blue')
plt.title('Bar plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid()
plt.show()
```

```python
#Giving colors to each bar
plt.bar(days,sales,color = ['red','green','blue','yellow','orange','purple'])
plt.title('Bar plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid()
plt.show()
```

```python
# Changing to horizontal plot. Use barh
plt.barh(days,sales,color = ['red','green','blue','yellow','orange','purple'])
plt.title('Horizontal Bar plot of Days and Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Days')
plt.grid()
plt.show()
```

```python
# Changing size of grid
plt.barh(days,sales,color = ['red','green','blue','yellow','orange','purple'])
plt.title('Horizontal Bar plot of Days and Sales')
plt.xlabel('Sales Amount')
plt.ylabel('Days')
plt.grid(True)
plt.show()
```

```python
# SCATTER PLOTS
roll_no = [101,102,103,104,105,106,107,108,109,110]
marks = [10,18,34,90,43,67,21,67,45,50]
plt.scatter(roll_no,marks, color = 'red',marker='*',s=200) # s is used to chnage the size of marker
plt.title('Scatter plot of Roll No and Marks', color='red',fontsize=20)
plt.xlabel('Roll Number of students')
plt.ylabel('Marks of the. students')
plt.grid(True)
plt.show()
```

```python
# create multiple lines in a single line plot

days = ['Monday', 'Tuesday', 'Wednesday', 'Thursday','Friday','Saturday']
sale1 = [50,60,40,90,35,84]
sale2 = [78,39,23,80,50,47]
sale3 = [56, 75,87,67,35,90]
plt.plot(days,sale1, color='green',marker='*',label='Sale1')
plt.plot(days,sale2,color='purple',marker='D'label='Sale2')
plt.plot(days,sale3,color='red',marker='v'label='Sale2')
plt.title('Line plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.show()
```

```python
# Adding labeel to the line plot
plt.plot(days,sale1, color='green',marker='*',label='Sale1')
plt.plot(days,sale2,color='purple',marker='D',label='Sale2')
plt.plot(days,sale3,color='red',marker='v',label='Sale2')
plt.title('Line plot of Days and Sales')
plt.xlabel('Days')
plt.ylabel('Sales Amount')
plt.grid(True)
plt.legend()  # The legend is used to ouput the labels
plt.show()
```

```python
# SCATTER PLOTS
roll_no = [101,102,103,104,105,106,107,108,109,110]
maths_marks = [10,18,34,90,43,67,21,67,45,50]
science_marks = [34,45,67,54,89,34,50,12,45,85]
plt.scatter(roll_no,maths_marks, color = 'red',marker='*',s=120,label='Maths') # s is used to chnage the size of marker
plt.scatter(roll_no,science_marks, color = 'blue',marker='v',s=120,label='Science') # s is used to chnage the size of marker
plt.title('Scatter plot of Roll No and Marks', color='purple',fontsize=20)
plt.xlabel('Roll Number of students')
plt.ylabel('Marks of the. students')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# STACKED BAR PLOTS
categories = ['A','B','C','D','E']
value1 = [10,20,30,40,50]
value2 = [5, 10, 15, 20, 25]
plt.bar(categories,value1,color='red',label='Value1')
plt.bar(categories,value2,color='blue',label='Value2',bottom=value1)
plt.title('Stacked Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# STACKED HORIZONTAL BAR PLOTS
categories = ['A','B','C','D','E']
value1 = [10,20,30,40,50]
value2 = [5, 10, 15, 20, 25]
plt.barh(categories,value1,color='red',label='Value1')
plt.barh(categories,value2,color='blue',label='Value2')
plt.title('Stacked Bar Plot')
plt.xlabel('Categories')
plt.ylabel('Values')
plt.grid(True)
plt.legend()
plt.show()
```

```python
# STACKED HORIZONTAL BAR PLOTS
categories = ['Children 0-18','Young Adult 26-40','Medium Adult 41-50','Adult 51-70','Senior Citizens']
value1 = [10,20,30,45,35]
plt.pie(value1,labels=categories,colors=['red','green','cyan','blue','gold'],autopct='$%0.2f%%') # autopct is used to in how much
plt.title('Pie Chart')
plt.show()
```

```python
# To explode the piechart
```

```python
plt.pie(value1,labels=categories,colors=['red','green','cyan','blue','gold'],autopct='$%0.2f%%', explode=[0.3,0,0.2,0.1,0]) # autopct is used to in how much
plt.title('Pie Chart')
plt.show()
```

```python
# To explode the piechart in 3D
```

```python
plt.pie(value1,labels=categories,colors=['red','green','cyan','blue','gold'],autopct='$%0.2f%%', explode=[0.3,0,0.2,0.1,0], shadow=True) # autopct is used to give percentage, shadow is 3D
plt.title('Pie Chart')
plt.show()
```

```python
# HISTOGRAM --> To compare continuous values/data
# continous is like height 1.7, 1.9, 1.78
# Discrete data - Number of bedroom in a house

age = [34, 13, 45,2, 67,23,89,67, 65, 53, 57, 12, 39,43, 61, 64, 90,20, 29]
plt.hist(age,color='blue',edgecolor='red',bins=5) #bins represent the number of ranges to be used
plt.title('Histogram to show the distribution of ages')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

```python
# HBOX PLOTS --> How to check
data = [1,2,3,4,5,6,6,21,18,21,15,21,6,7,8,9,10] # Box represents where  most part of the data lies
sns.boxplot(data)
plt.show()
```

```python
# HEAT MAPS --> Used  to check correlation between the columns
data = [[1,2,3],[15,9,26],[7,8,9]]
sns.heatmap(data,annot=True)
plt.show()
```
